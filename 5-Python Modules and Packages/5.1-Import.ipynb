### Importing Modules in Python: Modules and Packages


- In Python, modules and packages help organize and reuseb code. Here's a comprehensive guide on how to import them.


```python
import math
math.sqrt(16)  # This will import the math module and calculate the square root of 16
```

```python
from math import sqrt,pi
print(sqrt(16))  # This will directly use the sqrt function from the math module
print(sqrt(25))  # This will also directly use the sqrt function from the math module
print(pi)  # This will directly use the pi constant from the math module
```

```python
import numpy as np
np.array([1, 2, 3])  # This will import the numpy module and create a numpy array
```

```python
from math import *
print(sqrt(16))  # This will directly use the sqrt function from the math module
print(sqrt(25))  # This will also directly use the sqrt function from the math module
print(pi)  # This will directly use the pi constant from the math module
```

```python
from package.maths import addition
addition(2, 3)  # This will import the addition function from the package.maths module and add 2 and 3
```

```python
from package import maths
print(maths.addition(2,3))
print(maths.subtraction(5,2)) # This will import the maths module from the package and perform addition and subtraction
print(maths.multiplication(3,4)) # This will import the multiplication function from the package.maths module and multiply 3 and 4 
```

### Conclusion

- Importing modules and packages in python allows you to organize your code, reuse functionalities, and keep your projects clean and manageable. By understanding how to import modules,specific functions, and use relative imports within packages, you can structure your Python applications more effectively.
