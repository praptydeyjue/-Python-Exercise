### Introduction to Function


#### Definition:

- A Function is a block of code that performs a specific task. Functions help in organizing code, reusing code, and improving readability.


```python
## Syntax 
def function_name(parameters):
    """ Docstring for the function""" 
    # Function body
    # Code to be executed
    return expression
```

```python
## Why Functions?
num =24
if num % 2 == 0:
    print("The number is even")
else:
    print("The number is odd")
```

```python
def even_or_odd(num):
    """This function finds even or odd"""
    if num % 2 == 0:
        print("The number is even")
    else:
        print("The number is odd")
```

```python
# Calling the function
even_or_odd(24)
```

```python
## Function with multiple parameters
def add(a,b):
    return a+b  ## c = a+b  return c

addition=add(10,20)
print(addition)
```

```python
## Default parameters
def greet(name):
    print(f"Hello {name} welcome to the paradise")

greet("Anu")
```

```python
def greet(name="Tushar"):
    print(f"Hello {name} welcome to the paradise")

greet()
greet("Anu") ## This will overwrite the default parameter
```

```python
## Variable-Length Arguments
## Positional And Keyword Arguments
def print_number(*Tushar):
    for num in Tushar:
        print(num)

print_number(1, 2, 3, 4, 5,"Tushar")
```

```python
def print_number(*args):
    for num in args:
        print(num)

print_number(1, 2, 3, 4, 5,"Anu")
```

```python
## Keyword Arguments:
def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}:{value}")

print_details(name="Tushar", age=30, city="kolkata")
print_details(name="Anu", age=25, Country="India")
```

```python
def print_details(*args, **kwargs):
    for value in args:
        print(f"Positaional argument:{value}")
    
    for key, value in kwargs.items():
        print(f"{key}:{value}")
        
print_details(1,2,3,"anu", name="Tushar", age=30, city="kolkata")
```

```python
## Return statements
def multiply(a,b):
    return a*b

multiply(10, 20)
```

```python
## Return mupliple parameters
def calculate(a, b):
    return a+b, a-b, a*b, a/b
calculate(10, 20)
```
