#### 1. Celsius To Fahrenheit

- Problem Description:
  You are given a temperature in Celsius. Your task is to convert it to Fahrenheit and return the result.
- Formula:
  To convert Celsius to Fahrenheit, use the Formula:
  F = (9/5\*C) + 32


```python
## Function to convert temperature from celsius to fahrenheit
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

num = int(input("Enter the temperature in celsius: "))
result=celsius_to_fahrenheit(num)
print(f"The temperature {num} in Fahrenheit is: {result}")
```

#### 2. Area of a Rectangle

- Problem Description :
  You are given the length and breadth of a rectangle . Your task is to compute and return the area of the rectangle.
- Formula :
  To calculate the area of a rectangle:
  Area = length\*breadth


```python
### Function to use area of a rectangle
def area_of_rectangle(length, breadth):
    return length * breadth

length = float(input("Enter the length of the rectangle: "))
breadth = float(input("Enter the breadth of the rectangle: "))
result_area = area_of_rectangle(length, breadth)
print(f"The area of the rectangle with length {length} and breadth {breadth} is: {result_area}")
```

#### 3. Distance Covered by a Vehicle

- Problem Description:
  You are given the speed of a vehicle and the time it has traveled. Your task is to compute and return the distance traveled by the vehicle.
- Formula:
  To calculate the distance traveled by a vehicle:
  Distance=SpeedÃ—Time


```python
### Function to use Distance Covered by a Vehicle
def distance_covered(speed, time):
    return speed * time 
speed = float(input("Enter the speed of the vehicle (in km/h): "))
time = float(input("Enter the time traveled (in hours): ")) 
result_distance = distance_covered(speed, time)
print(f"The distance covered by the vehicle at speed {speed} km/h for {time} hours is: {result_distance} km")
```

#### 4. Number of Rounds of Lift

- Problem Description:
  You are given n, the total number of people, and capacity, the maximum number of people the lift can carry at a time. All people want to go from the ground floor to the top floor. Your task is to calculate the number of rounds the lift has to make to transport all the people to the top floor.
- Input:
  Two integers, n and capacity, where n is the total number of people, and capacity is the maximum number of people the lift can carry in one round.


```python
## Function to use Number of Rounds of lift
def calculate_lift_rounds(n,capacity):
    return (n + capacity - 1) // capacity ### This formula calculates the number of rounds needed by the lift
n = int(input("Enter the total number of people: "))
capacity = int(input("Enter the capacity of the lift: "))
result_lift_rounds = calculate_lift_rounds(n, capacity)
print(f"The number of rounds the lift needs to make for {n} people with a capacity of {capacity} is: {result_lift_rounds}")
```

#### 5.Line Equation

- Problem Description:
  You are given the slope m and the y-intercept b of a line, along with a value x. Your task is to calculate and return the value of y using the equation of a line in slope-intercept form: y=mx+b
- Input:
  Three floating-point numbers: slope, intercept, and x.


```python
## Function to use line Equation.
def calculate_y(slope, x, intercept):
    return slope * x + intercept
slope = float(input("Enter the slope (m): "))
x = float(input("Enter the x-coordinate(x): "))
intercept = float(input("Enter the y-intercept (b): "))
result_y = calculate_y(slope, x, intercept)
print(f"The value of y for the line equation y = {slope}x + {intercept} at x = {x} is: {result_y}")
```
