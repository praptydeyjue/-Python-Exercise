```python
# While Loops
# The while loop continues to execute as long as the condition is True.

count=0

while count<5:
    print(count)
    count =count+1
```

```python
## Loop Control statements
## Break
## The break statement exits the loop  permaturely
## break statement is used to exit the loop

for i in range(10):
    if i ==5:
        break
    print(i)

```

```python
## Continue Statement
## The continue statement skips the current iteration of a loop and continues with the next iteration.

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
#pass statement
## The pass statement is a null statement. It does nothing.

for i in range(5):
    if i==3:
        pass
    print(i)
```

```python
# Nested Loops
## a loop inside a loop

for i in range(3):
    for j in range(2):
        print(f"i:{i} and j:{j}")
```

```python
## Examples - Calculate the sum of first n natural numbers using a while loop and  for loop
## while loop  

n=10
sum=0
count=1

while count<=n:
    sum= sum+count
    count = count+1
    
print("Sum of first 10 natural numbers is :",sum)
```

```python
n=10
sum=0

for i in range(n+1):
    sum+=i
print("Sum of first 10 natural numbers is :",sum)
```

```python
##  Example - Prime numbers between 1 to 100

for num in range(1,101):
    if num > 1:
        for i in range(2,num):
            #print(i, num)
            if num%i==0:
                break
        else:
            print("Prime Number:",num)
```

### Conclusion:

Loops are a powerful construct in python that allow you to execute a block of code multiple times.By understanding and using for and while loops,along with loop control statements like break and continue, and pass, you can handle a wide range of programming tasks efficiently.
