### Introduction To Tuple

- Tuples are used to store multiple items in a single variable.
- Tuple is one of 4 built-in data types in Python used to store collections of data, the other 3 are List, Set, and Dictionary, all with different qualities and usage.
- A tuple is a collection which is ordered and unchangeable.
- Tuples are written with round brackets ( ).

```python
numbers = tuple([1, 2, 3, 4, 5])
print(numbers)
```

```python
list((1, 2, 3, 4, 5,6))
```

```python
mixed_tuple = (1,"Hello World", 3.14, True)
print(mixed_tuple)
```

```python
## Accessing Tuple Elements:
numbers
```

```python
print(numbers[0]) # Accessing first element
print(numbers[-1]) # Accessing last element
print(numbers[0:4]) # Slicing from index 1 to 3 (4 is excluded)
print(numbers[::-1]) # Reversing the tuple
print(numbers[1:4:2]) # Slicing with step 2 (1, 3)
```

```python
##Tuple operations:
concatenated_tuple = numbers + mixed_tuple
print(concatenated_tuple) # Concatenation of two tuples
```

```python
mixed_tuple1 = mixed_tuple * 3 
print("Original tuple:", mixed_tuple)
print("Mixed tuple:", mixed_tuple1) # Repeating the tuple 3 times
print("Length of mixed tuple:", len(mixed_tuple1)) # Length of the tuple
```

```python
numbers*3 # Repeating the tuple 3 times
```

```python
##Immutable Nature of Tuples:
##Tuples are immutabele, menaing their elements cannot be changed once assigned.

#List is mutable, meaning their elements can be changed once assigned.
lst = [1, 2, 3, 4, 5]
print(lst)

lst[1]= "Tushar"
print(lst)
```

```python
#Tuple 
numbers[1] = "Tushar" # This will raise an error because tuples are immutable   
```

```python
##Tuble Methods:
num=(1, 3, 7, 8, 7, 4, 6, 8, 5)# Creating a tuple
print(num.count(7))# Counting the number of occurrences of 1 in the tuple
print(num.index(5))# Finding the index of the first occurrence of 5 in the tuple. Searches the tuple for a specified value and returns the position of where it was found
print(num[2:5])# Slicing the tuple to get elements from index 2 to 4
```

```python
##Packing and unpacking of tuples
##Packing of tuples:
packed_tuple = 1,"Hello",3.14,
print(packed_tuple) 
```

```python
## Unpacking of tuples:
a,b,c = packed_tuple # Unpacking the tuple into three variables
print(a) 
print(b)
print(c)  
```

```python
##unpacking with
numbers = (1, 2, 3, 4, 5)
first, *middle,last = numbers # Unpacking the first, middle and last elements of the tuple
print(first) # 1
print(middle) # [2, 3, 4]
print(last) # 5
```

```python
##Nested Tuples:
##Nested lists:
lst = [[1,2,3,4],[6,7,8,9],(1,"Hello",3.14,"c")]
print(lst[0][0:3]) # Accessing the first three elements of the first list
print(lst[1][0:3]) # Accessing the first three elements of the second list
print(lst[2][0:2]) # Accessing the first three elements of the third list
```

```python
nested_tuple = ((1,2,3,4),("a","b","c","d"),(True,False))
#access the elements inside a tuple
print(nested_tuple[0]) # Accessing the first tuple
print(nested_tuple[1][3]) # Accessing the last element of the second tuple
print(nested_tuple[2][1]) # Accessing the second element of the third tuple


```

```python
##Iterating over a tuple:
for sub_tuple in nested_tuple:
    for element in sub_tuple:
        print(element, end= " ")# Printing each element on the same line
    print() # Printing a new line after each sub-tuple
```
