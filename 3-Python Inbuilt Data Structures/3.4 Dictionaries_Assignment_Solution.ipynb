#### Assignment 6 : Meraging Dictionaries


```python
#Create two dictionaries:One with keys as the first 5 positive integers and values as their squares, and another with keys as the next 5 positive integers and values as their squares. Merage these dictionaries into a single dictionary and print it.
d1 ={i: i**2 for i in range(1,6)}
d2 = {i: i**2 for i in range(6,11)}
d1.update(d2)
print(d1)
```

```python
#Another way to merge two dictionaries
#Create two dictionaries:One with keys as the first 5 positive integers and values as their squares, and another with keys as the next 5 positive integers and values as their squares. Merage these dictionaries into a single dictionary and print it.
d1 ={i: i**2 for i in range(1, 6)} 
d2 ={i: i**2 for i in range(6,11)}
d1 = {**d1, **d2} ### Merge 2 dictionaries into one
print(d1)
```

#### Assignment 11 : Dicitionay Filtering


```python
#Create a dictionary with the first 10 positive integers as keys and their squares as values. Create a new dictionary containing only the key-value pairs where the key is even. Print the new dictionary.
d ={i:i**2 for i in range(1,11)}
even_dict = { k:v for k,v in d.items() if k%2==0}
print(even_dict)
```

#### Assignment 14: Counting with Dictionaries


```python
#Write a function that takes a string and returns a dictionary containing the count of each character in the string. Print the dictionary.
def count_characters(string):
    count_dict = {}
    for char in string:
        count_dict[char] = count_dict.get(char,0) +1 
    return count_dict 

string = "TUSHARDEBNATH"
print(count_characters(string))
```
