##### List Comprehension

Basics Syantax [expression for item in iterable]

with conditional logic [expression for item in iterable if condition]

Nested List Comprehension [expression for item1 in iterable1 for item2 in iterable2]


```python
## Basic List Comprehension
sqaure = [num**2 for num in range(20)]
print(sqaure)
```

```python
## List Comprehension with if condition
lst = []
for i in range(10):
    if i%2==0:
        lst.append(i)
print(lst)
```

```python
even_num =[i for i in range(100) if i%2==0 ]
print(even_num)
```

```python
## Nested list Comprehension
lst1=[1,2,3]
lst2=['a','b','c','d']

pair=[[i,j] for i in lst1 for j in lst2]
print(pair)
```

```python
## List Comprehension with funcaiton call

words = ["hello", "world", "python", "programming"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
## Copy a list
a = [1, 2, 3, 4, 5]
b = a.copy()
print(a)
print(b)
```

```python
## Copy a list with append
a = [1, 2, 3, 4, 5]
b = a.copy()
b.append(6) #Append 6 to list b
print(a)
print(b)
```

```python
## Copy a list with Mixed Data Types(Shallow Copy)
c=[1, 'two',3.0,[4,5],{'key':'value'}]
d=c.copy()

d[3].append(6) #Append 6 to list inside list b

print("C:",c)
print("D:",d)
```

```python
#Freeze the list, and make it unchangeable
frozen_set = frozenset(range(1,6)) 
print(frozen_set)
```

```python
mylist = ['apple', 'banana', 'cherry']
x = frozenset(mylist)
print(x)
```
