## Sets

Sets are a built-in data type in python used to store collections of unique items.They are unordered, meaning that the elements do not follow a specific order, and they do not allow duplicate elements.Sets are useful for membership tests, eliminating duplicate entries, and performing mathematical set operations like union,intersection, difference, and symmetric difference.

- Note: Set items are unchangeable, but you can remove items and add new items. A set is a collection which is unordered, unchangeable\*, and unindexed.Sets are used to store multiple items in a single variable.

- Sets are written with curly brackets.

```python
## Basic sets Operations
##Adding and Removing elements from a set
set1 = {1,2,3,4,5,6} 
set1.add(7)  # add an element to the set
print(set1)

```

```python
set1.remove(3)  # remove an element from the set
print(set1)
```

```python
set1.remove(10)  # remove an element from the set
print(set1)
```

```python
set1.discard(11)  # remove an element from the set
print(set1)# discard does not raise an error if the element is not found in the set
set1.discard(7)  # remove an element from the set
print(set1)  # discard does not raise an error if the element is not found in the set
```

```python
##pop method
removed_element = set1.pop()  # remove and return an arbitrary element from the set
print(removed_element)
print(set1)
```

```python
set1.clear()  # remove all elements from the set
print(set1) # set1 is now an empty set
```

```python
# set Memebership test
# check if an element is in the set
my_set1 = {1, 2, 3, 4, 5}
print(3 in my_set1) # check if an element is in the set
print(10 in my_set1) # check if an element is in the set
print(4 not in my_set1) # check if an element is not in the set
```

```python
## Set Mathematics Operations
set2 = {1,2,3,4,5,6}
set3 = {4,5,6,7,8,9}

##Union
union_set = set2.union(set3) # union of two sets 
print(union_set)

union_set = set2 | set3 # union of two sets using the | operator
print(union_set)
```

```python
##Intersection
intersection_set = set2.intersection(set3) # intersection of two sets
print(intersection_set) # intersection of two sets

intersection_set = set2 & set3 # intersection of two sets using the & operator  
print(intersection_set) # intersection of two sets using the & operator
```

```python
##Difference
difference_set = set2.difference(set3) # difference of two sets
print(difference_set)
difference_set = set2 - set3 # difference of two sets
print(difference_set)
```

```python
##Difference
difference_set = set3.difference(set2) # difference of two sets
print(difference_set)
difference_set = set3 - set2 # difference of two sets using the - operator
print(difference_set)
```

```python
##Symmetric Difference
sym_diff=set2.symmetric_difference(set3) #symmetric difference of two sets
print(sym_diff)
sym_diff=set2^set3 #symmetric difference of two sets using the ^ operator
print(sym_diff)


```

```python
##Subset and Superset

a={1,2,3,4,5,6}
b={1,2,3}
#subset
print(a.issubset(b)) # check if a is a subset of b
print(b.issubset(a)) # check if b is a subset of a
print(a<=b) # check if a is a subset of b using the <= operator
print(b<=a) # check if b is a subset of a using the <= operator

```

```python
#Superset
print(a.issuperset(b)) # check if a is a superset of b
print(b.issuperset(a)) # check if b is a superset of a 
print(a>=b) # check if a is a superset of b using the >= operator
print (b>=a) # check if b is a superset of a using the >= operator
```

```python
### Disjoint Sets 


## Return True if no items in set x is present in set y
## The isdisjoint() method returns True if none of the items are present in both sets, otherwise it returns False.
x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "facebook"}

z = x.isdisjoint(y)

print(z)
##What if no items are present in both sets?
##Return False if one or more items are present in both sets:

x = {"apple", "banana", "cherry"}
y = {"google", "microsoft", "apple"}

z = x.isdisjoint(y)

print(z)
```

```python
## Counting Unique words in text
text = "In this tutorial we are discussing about set in python"
words = text.split() # split the text into words

##convert list of words to set to get unique words
unique_words = set(words) # convert list of words to set to get unique words
print(unique_words)
print(len(unique_words))
```
