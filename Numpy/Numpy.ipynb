```python
#indexing & slicing
#1D
num=np.array([1,3,5,7,9,4,2,7,9,8])
print(num[0])
print(num[1])
print(num[-2])
print(num[:3])
print(num[2:])
print(num[2:4])
print(num[-4:-1])
print(num[:5:2])
print(num[-8:-2:2])
```

```python
#2D
num2=np.array([[1,2,3,4,5,6],[9,8,7,11,13,21]])
print(num2[0,1])
print(num2[1,2])
print(num2[1,-2])
print(num2[1,:3])
print(num2[0,2:])
print(num2[1,2:4])
print(num2[0:2,-4:-1])
print(num2[1,:5:2])

```
