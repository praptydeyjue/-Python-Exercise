## Arithmetic Operators:


```python
##Arithmetic Operations

a=10
b=5

addition_result = a+b #addition
subtraction_result = a-b #subtraction
mulplication_result = a*b #multiplication
division_result = a/b #division
modulus_result = a%b #modulus
floor_division_result = a//b # floor division
exponent_result = a**b #exponentiation

print("Addition Result: ",addition_result)
print("Subtraction Result:", subtraction_result)
print("Multiplication Result:", mulplication_result)
print("Division Result:", division_result)
print("Modulus Result:", modulus_result)
print("Floor Division Result:", floor_division_result)
print("Exponent Result:", exponent_result)




```

## Comparison Operators:


```python
##Comparison Operators

#Equal
print("Equal:")
print(2==2)
print(2==1)

#Not Equal
print("Not Equal:")
print(3!=2)
print(3!=3)

#Greater Than
print("Greater Than:")
print(2>1)
print(1>2)

#Less Than
print("Less Than:")
print(2<3)
print(3<2)

#Greater Than or Equal to
print("Greater Than or Equal to:")
print(2>=2)
print(2>=1)

#Less Than or Equal to
print("Less Than or Equal to:")
print(2<=3)
print(2<=1)
```

```python
#Greter than or equal to ">="
number3 = int(input("Enter a number: "))
number4 = int(input("Enter another number: "))

if number3 >= number4:
    print(number3, "is greater than or equal to", number4)
    
else:
    print(number3, "is less than", number4)
```

## Logical Operators:


```python
# And operation : and operation is true if both the operands are true otherwise it is false
X=True
Y=True

result= X and Y
print(result)
```

```python
X=True
Y=False

result= X and Y
print(result)
```

```python
#OR operation : OR operation is true if any one of the operands is true otherwise it is false

X=True
Y=True

result= X or Y
print(result)
```

```python
X=True
Y=False

result= X or Y
print(result)
```

```python
X=False
Y=False

result= X or Y
print(result)
```

```python
#Not operation : Not operation is true if the operand is false and vice versa

X=True
not X
```

```python
X=False
not X
```
