```python
#understanding variables types
##Pyhton is a dynamically typed, type of a variable is determined at runtime

age=32#integer
height=6.1 #float
name="DebNath" #string
is_student=True #boolean

print(type(age))
print(type(height))
print(type(name))
print(type(is_student))
```
